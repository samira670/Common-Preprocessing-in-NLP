```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.probability import FreqDist
import string
from collections import Counter
import matplotlib.pyplot as plt

# Download necessary components for processing
nltk.download('punkt')  # Used for tokenizing sentences and words
nltk.download('stopwords')  # Used for retrieving a list of stopwords
nltk.download('wordnet')  # Used for lemmatization
nltk.download('averaged_perceptron_tagger')  # Used for part-of-speech tagging

# Example text with repetition to illustrate poetic rhythm
text = "The sun shines, the sun sets; the moon glows, the moon fades; the stars twinkle, the stars die, but love, oh love, it remains, it sustains."

# Sentence Tokenization: Splits the text into a list of sentences.
sentences = sent_tokenize(text)
print("Sentence Tokenization:")
print(sentences)

# Word Tokenization: Splits sentences into individual words.
words = word_tokenize(text)
print("\nWord Tokenization:")
print(words)

# Cleaning the data: Removes stopwords and punctuation from the list of words.
stop_words = set(stopwords.words('english') + list(string.punctuation))
cleaned_words = [word for word in words if word.lower() not in stop_words]
print("\nCleaned Data:")
print(cleaned_words)

# Frequency Distribution using Counter
word_freq_counter = Counter(cleaned_words)
print("\nWord Frequency using Counter:")
print(word_freq_counter)

# Frequency Distribution using FreqDist
word_freq_fdist = FreqDist(cleaned_words)
print("\nWord Frequency using FreqDist:")
print(word_freq_fdist)

# Plotting the frequency of words using FreqDist
plt.figure(figsize=(10, 5))
word_freq_fdist.plot(30, cumulative=False, title='Frequency Distribution of Words using FreqDist')
plt.show()

# Stemming: Reduces words to their base form by cutting off the ends of words.
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in cleaned_words]
print("\nStemmed Words:")
print(stemmed_words)

# Lemmatization: Converting words into their dictionary form by considering their part of speech.
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in cleaned_words]
print("\nLemmatized Words:")
print(lemmatized_words)

# Part-of-Speech Tagging: Assigns a part of speech to each word in the cleaned list.
pos_tags = pos_tag(cleaned_words)
print("\nPart-of-Speech Tags:")
print(pos_tags)
```
